First we upload the libraries that will be used: stats, numpy, matplotlib


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

Next we will learn how to compute basic statistical estimators: mean, variance, median

```python
x = [7.6, 5.8, 8.0, 6.9, 7.2, 7.5, 6.4, 8.1, 6.3, 7.0] # 10 measurements of a variable
s = stats.describe(x)
mu = s.mean # mean
var = s.variance # variance
sig = np.sqrt(var) # standard deviation
median = np.median(x) # median
print('Mean =',mu)
print('Variance =',var)
print('Standard deviation =',sig)
print('Median =',median)
```

These estimators can also be computed with the numpy library

```python
x=np.array(x)
vm=np.mean(x)
var=np.var(x,ddof=1)
desv=np.std(x,ddof=1)
print('Mean =',vm)
print('Variance =',var)
print('Standard deviation =',desv)
```

## Binomial distribution 



The binomial distribution applies to random processes with **two possible outcomes** that have probabilities $p$ and $1-p$.

If we perform $N$ independent trials, where the probability of "success" in each trial is $p$, then the probability of obtaining exactly $n$ successes is:

$P_{Binomial}(n) = \frac{N!}{n! (N-n)!} p^n (1-p)^{N-n}$

The mean and variance of this distribution are $\overline{n} = pN$ and $\text{Var}(n) = Np(1-p)$, respectively.

Below is an example of a binomial distribution with $N = 10$ and $p = 0.2$:



```python
import matplotlib.pyplot as plt
#matplotlib inline
N = 10 # number of trials
p = 0.2 # probability of success in each trial 
n = np.arange(N+1) # number of succesful trials for which we compute the probability
y = stats.binom.pmf(n,N,p) # binomial probability distribution
mu = N*p # mean value of the binomial distribution
xmin,xmax = 0.,float(N)
ymin,ymax = 0.,0.4
plt.figure()
plt.plot(x,y,marker='o',markersize=5,color='black')
plt.plot([mu,mu],[ymin,ymax],color='black',linestyle='dotted')
plt.xlabel(r'$n$')
plt.ylabel(r'$P_{\rm Binomial}(n)$')
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
plt.savefig('binomial.png',format="png")
plt.show()
plt.close()
```

# Poisson Distribution

The Poisson distribution applies to **discrete random processes where events are counted** in a fixed interval of time or space. If the mean number of expected events is $\mu$, the probability of observing $n$ events is:

$P_{\rm Poisson}(n) = \frac{\mu^n \, e^{-\mu}}{n!}$

The mean and variance of this distribution are equal: $\overline{n} = {\rm Var}(n) = \mu$.

Below is an example of a Poisson distribution with $\mu = 5$:

```python
mu = 5. # mean value of the expected number of events
nmax = 15 # maximum number of events
x = np.arange(nmax+1) # number of events for which we compute the probability
y = stats.poisson.pmf(x,mu) # poisson probability distribution for a mean mu
xmin,xmax = 0.,float(nmax)
ymin,ymax = 0.,0.25
plt.plot(x,y,marker='o',markersize=5,color='black')
plt.plot([mu,mu],[ymin,ymax],color='black',linestyle='dotted')
plt.xlabel(r'$n$')
plt.ylabel(r'$P_{\rm Poisson}(n)$')
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
```

# Gaussian Distribution

The **Gaussian or normal distribution** for a variable $x$, with mean $\mu$ and standard deviation $\sigma$, is:

$P_{\rm Gaussian}(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x-\mu}{\sigma} \right)^2}$

Below is an example of a Gaussian distribution with $\mu = 2$ and $\sigma = 0.5$:

```python
mu = 2. # mean value
sig = 0.5 # standard deviation
xmin,xmax,nx = 0.,4.,101 # range and number of points where we evaluate the gaussian distribution
x = np.linspace(xmin,xmax,nx) # values for which we compute the gaussian distribution
y = stats.norm.pdf(x,mu,sig) # gaussian probability distribution with mean=mu and standard deviation=sig
ymin,ymax = 0.,1.
plt.plot(x,y,color='black')
for i in range(-3,4):
  x1 = mu + i*sig
  plt.plot([x1,x1],[ymin,ymax],color='black',linestyle='dotted')
plt.xlabel(r'$x$')
plt.ylabel(r'$P_{\rm Gaussian}(x)$')
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
```
